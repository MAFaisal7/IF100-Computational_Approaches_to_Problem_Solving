```python
new_tx = input("Please enter your purchase history: ")
currencies = []
amounts = []
new_txs = new_tx.split(";")
for j in new_txs:
    new_c = j.split(":")[0]
    new_a = float(j.split(":")[1])
    if new_c in currencies:
        ind = currencies.index(new_c)
        if new_a < 0:
            current = amounts[ind]
            if current >= -new_a:
                amounts[ind] += new_a
                print(str(-new_a) + " " + new_c + " sold") 
            else:
                print("Not enough " + new_c)
        else:
            print(str(new_a) + " " + new_c + " bought") 
            amounts[ind] += new_a
    else:
        if new_a < 0:
            print("You don't have " + new_c)
        else:
            print(str(new_a) + " " + new_c + " bought") 
            currencies.append(new_c)
            amounts.append(new_a)

cur_type = input("Please enter the currency type: ")
fiat_price = input("Please enter prices: ")
fiat_prices = fiat_price.split(";")
values = []
converted_currencies = []
for price in fiat_prices:
    cross = price.split(":")[0]
    value = float(price.split(":")[1])
    lhs = cross.split("_")[0]
    rhs = cross.split("_")[1]
    if rhs == cur_type:
        if lhs not in converted_currencies:
            values.append(value)
            converted_currencies.append(lhs)
    elif lhs == cur_type:
        if rhs not in converted_currencies:
            values.append(1/value)
            converted_currencies.append(rhs)
total = 0
for c in currencies:
    a = amounts[currencies.index(c)]
    if c == cur_type:
        total += a
    else:
        price = values[converted_currencies.index(c)]
        total += (a*price)
print("You have " + format(total,".2f") + " " + cur_type + "(s).")
```
